Создайте в ваших данных о транзакциях признак `QuantityCanceled`. Этот признак будет указывать на количество отменённого впоследствии товара для каждой транзакции, на которую найдётся противоположная ей транзакция с возвратом. Для транзакций, для которых не было возвратов, признак будет равен 0. Для транзакций, для которых указано отрицательное количество товара, но в данных отсутствует контрагент, признак будет пустым.

В качестве вспомогательного инструмента мы подготовили для вас функцию `get_quantity_canceled()`. Эта функция принимает на вход таблицу с транзакциями и возвращает объект `Series` — столбец, в котором указано количество отменённого впоследствии товара для каждой транзакции. Если транзакция не имеет контрагента, этот признак помечается как `NaN`.

Отметим, что эта функция не учитывает сложный случай, когда количество отменённого товара в транзакции-возврате больше, чем количество товара, которое указано в любой из отдельных транзакций на покупку (это случай, когда клиент сделал несколько заказов одного и того же товара, а потом оформил возврат на все товары разом). При желании вы можете самостоятельно модифицировать функцию для предобработки, чтобы учесть этот пограничный случай.

*Осторожно! Поиск отменённых транзакций занимает много времени из-за большого количества строк в таблице. Выполнение следующего кода может занять до 20 минут в зависимости от мощности компьютера.*

Create a 'QuantityCanceled' feature in your transaction data. This feature will indicate the quantity of items subsequently canceled for each transaction that has a counter-transaction with a refund. For transactions with no refunds, the feature will be 0. For transactions with a negative quantity but no counterparty in the data, the feature will be empty.

As a helper, we've prepared the 'get_quantity_canceled()' function for you. This function takes a table of transactions as input and returns a 'Series' object—a column that indicates the quantity of items subsequently canceled for each transaction. If a transaction has no counterparty, this feature is marked as 'NaN'.

Please note that this function does not account for the complex case where the quantity of the cancelled item in the return transaction is greater than the quantity specified in any of the individual purchase transactions (this is the case when a customer placed multiple orders for the same item and then issued a return for all of them at once). You can modify the preprocessing function to account for this edge case if desired.

*Caution! Searching for cancelled transactions takes a long time due to the large number of rows in the table. The following code may take up to 20 minutes to run, depending on your computer's performance.*

```python
def get_quantity_canceled(data):
    """Функция для создания признака количества отменённых заказов. 
    Функция принимает на вход таблицу и возвращает столбец, в котором указано количество отменённого впоследствии товара для кажой транзакции.
    Если транзакция с отрицательным количеством товара не имеет контрагента, данный признак помечается как NaN.

    Args:
        data (DataFrame): таблица с транзакциями

    Returns:
        Series: столбец с количеством отменённого товара
    """
    # Инициализируем нулями Series той же длины, что и столбцы таблицы
    # Initialize a Series with zeros and the same length as the table columns
    quantity_canceled = pd.Series(np.zeros(data.shape[0]), index=data.index)    
    negative_quantity = data[(data['Quantity'] < 0)].copy()
    for index, col in negative_quantity.iterrows():
        # Создаём DataFrame из всех контрагентов
        # Create a DataFrame from all counterparties
        df_test = data[(data['CustomerID'] == col['CustomerID']) &
                       (data['StockCode']  == col['StockCode']) & 
                       (data['InvoiceDate'] < col['InvoiceDate']) & 
                       (data['Quantity'] > 0)].copy()
        # Транзация-возврат не имеет контрагента — ничего не делаем
        # The return transaction has no counterparty - we do nothing
        if (df_test.shape[0] == 0): 
            # Помечаем столбец как пропуск
            quantity_canceled.loc[index] = np.nan
        # Транзакция-возврат имеет ровно одного контрагента
        # A return transaction has exactly one counterparty
        # Добавляем количество отменённого товара в столбец QuantityCanceled
        # Add the quantity of the cancelled item to the QuantityCanceled column
        elif (df_test.shape[0] == 1): 
            index_order = df_test.index[0]
            quantity_canceled.loc[index_order] = -col['Quantity']       
        # Транзакция-возврат имеет несколько контрагентов
        # The return transaction has multiple counterparties
        # Задаём количество отменённого товара в столбец QuantityCanceled для той транзакции на покупку,
        # в которой количество товара больше количества товаров в транзакции-возврате.
        # We set the quantity of the cancelled product in the Quantity Canceled column for that purchase transaction,
        # in which the quantity of the product is greater than the quantity of the product in the return transaction.
        elif (df_test.shape[0] > 1): 
            df_test.sort_index(axis=0 ,ascending=False, inplace = True)        
            for ind, val in df_test.iterrows():
                if val['Quantity'] < -col['Quantity']: 
                    continue
                quantity_canceled.loc[ind] = -col['Quantity']
                break    
    return quantity_canceled

data['QuantityCanceled'] = get_quantity_canceled(data)
```

### 2.3. Построение RFM-таблицы и поиск RFM-выбросов
### 2.3. Построение RFM-таблицы и поиск RFM-выбросов

Мы добрались до самой интересной части нашей задачи. Нам предстоит сформировать признаки, на основе которых мы будем производить сегментацию клиентов.

Для этого давайте познакомимся с очень популярным методом для анализа потребительской ценности под названием RFM. 

<center> <img src=https://miro.medium.com/max/1400/1*uYQjy9SUjW7iWHc2gGanQQ.png align="right" width="400"/> </center>

Метод заключается в группировке клиентов на основе следующих параметров:
* Recency (Давность) — давность последней покупки клиента;
* Frequency (Частота) — общее количество покупок клиента;
* Monetary Value (Денежная ценность) — сколько денег потратил клиент.


Суть RFM-анализа состоит в том, что мы разделяем всех клиентов на группы в зависимости от того, как давно они сделали последнюю покупку, как часто покупали и насколько большой была сумма их заказов. 

Например, вот так может выглядеть интерпретация кластеров для случая RF-сегментации (анализа на основе давности и частоты заказов клиента):

<img src=https://retailrocket.ru/wp-content/uploads/2017/06/rfm-1.png>

Задача маркетологов — вести клиента в зону лояльных.

Мы можем рассчитать RFM-характеристики для каждого из клиентов в нашем датасете и на их основе с помощью методов кластеризации построить подобные сегменты клиентов, привязанные к нашим данным.

Чтобы получить RFM-таблицу, нам необходимо сгруппировать данные по идентификаторам клиента и рассчитать следующие  агрегированные характеристики:

* Recency для i-го клиента рассчитывается как разница между датой и временем последнего заказа и точкой отсчёта, переведённая в дни:
    $$t_0-max(t_{i1}, t_{i2},..., t_{iM})$$

    где $t_{ij}$ — дата и время совершения i-ым клиентом своей j-ой покупки.

    В качестве точки отсчёта $t_0$ берём дату на один день «старше», чем все наши данные. Это будет 10 декабря 2011 года (в формате datetime — '2011-12-10 00:00:00').

* Frequency рассчитывается как общее количество уникальных заказов, которые совершил i-ый клиент.
* Monetary Value рассчитывается как общая сумма денег, которую i-ый клиент потратил на наши товары (с учётом возвратов).

Когда вы рассчитаете все характеристики, не забудьте дать столбцам результирующей таблицы соответствующие названия.

We've reached the most interesting part of our task. We need to develop the characteristics we'll use to segment our customers.

To do this, let's take a look at a very popular method for analyzing customer value called RFM.

<center> <img src=https://miro.medium.com/max/1400/1*uYQjy9SUjW7iWHc2gGanQQ.png align="right" width="400"/> </center>

This method involves grouping customers based on the following parameters:
* Recency — how long ago the customer last made a purchase;
* Frequency — the total number of purchases the customer made;
* Monetary Value — how much money the customer spent.

The essence of RFM analysis is that we divide all customers into groups based on how long ago they made their last purchase, how often they purchased, and the size of their order total.

For example, this is how cluster interpretation might look for RF segmentation (analysis based on the recency and frequency of a customer's orders):

<img src=https://retailrocket.ru/wp-content/uploads/2017/06/rfm-1.png>

Marketers' task is to guide customers into the loyal zone.

We can calculate RFM characteristics for each customer in our dataset and, based on these, use clustering methods to construct similar customer segments linked to our data.

To obtain an RFM table, we need to group the data by customer IDs and calculate the following aggregated characteristics:

* Recency for customer i is calculated as the difference between the date and time of the last order and the reference point, converted into days:
$$t_0-max(t_{i1}, t_{i2},..., t_{iM})$$

where $t_{ij}$ is the date and time of customer i's jth purchase.

For the reference point $t_0$, we take the date one day "older" than all of our data. This will be December 10, 2011 (in datetime format - '2011-12-10 00:00:00').

* Frequency is calculated as the total number of unique orders placed by customer i. * Monetary Value is calculated as the total amount of money the i-th customer spent on our products (including returns).

Once you've calculated all the characteristics, don't forget to name the columns in the resulting table accordingly.

```python
# Задаём точку отсчёта t0
# Set the reference point t0
t0 = pd.Timestamp('2011-12-10 00:00:00')

# Группируем данные по клиентам
# Grouping data by clients
rfm_table = (
    data_copy.groupby('CustomerID')
    .agg({
        'InvoiceDate': 'max',
        'InvoiceNo': 'nunique',
        'TotalPrice': 'sum'
    })
)

# Считаем Recency
# Calculating Recency
rfm_table['Recency'] = (t0 - rfm_table['InvoiceDate']).dt.days

# Переименовываем столбцы
# Переименовываем столбцы
rfm_table = rfm_table.rename(columns={
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
})

# Удаляем ненужную колонку
# Delete the unnecessary column
rfm_table = rfm_table.drop(columns='InvoiceDate')

rfm_table.head()
```

Что интересного здесь можно увидеть? Есть клиенты с аномально большим количеством сделанных заказов (более 100 штук), а также клиенты, общая стоимость заказов которых превышает 190 тысяч фунтов стерлингов.

Чем это плохо? Выбросы могут отрицательно сказаться на результатах работы методов кластеризации, неустойчивых к ним, например алгоритма KMeans, поэтому хотелось бы от них избавиться. Однако терять много ценных данных о клиентах тоже не хочется, поэтому ограничимся верхней границей соответствующей квантили уровня 0.95. Таким образом, мы удалим данные тех клиентов, для которых значение параметра Frequency или параметра Monetary выше, чем у 95 % клиентов.


What's interesting to see here? There are customers with an abnormally large number of orders (more than 100), as well as customers with a total order value exceeding £190,000.

Why is this bad? Outliers can negatively impact the results of clustering methods that are not robust to them, such as the KMeans algorithm, so we'd like to remove them. However, we also don't want to lose a lot of valuable customer data, so we'll limit the upper bound of the corresponding quantile to 0.95. This way, we'll remove data from customers for whom the Frequency or Monetary parameter value is higher than 95% of customers.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2))
])

# Применяем pipeline
# Applying the pipeline
X = rfm_table[['Recency','Frequency','Monetary']]

X_pca = pipeline.fit_transform(X)
```

```python
# На основе RFM-признаков кластеризуйте клиентов с помощью KMeans
# Cluster customers based on RFM features using KMeans
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# напишем функцию, чтобы рассчитать коэффициент силуэта
# Let's write a function to calculate the silhouette coefficient
def get_silhouette(cluster_num, X):
    km =  KMeans(n_clusters=cluster_num, random_state=42)
    km.fit(X)
    # подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
    # calculate the silhouette metric by passing the data and which clusters the objects belong to
    silhouette = silhouette_score(X, km.predict(X))
    return silhouette

# создадим пустой словарь, ключами будут коэффициент силуэта и количество кластеров
# create an empty dictionary, the keys will be the silhouette coefficient and the number of clusters
silhouette_res = {"silhouette": [], "cluster": []}

for cluster_num in range(3, 11):
    silhouette_res["silhouette"].append(get_silhouette(cluster_num, X_pca))
    silhouette_res["cluster"].append(cluster_num)
    
# сохраним в датафрейм значение силуэта и количество кластеров
# save the silhouette value and the number of clusters in the dataframe
silhouette_df = pd.DataFrame(silhouette_res)

# установим стиль для визуализиции
# set the style for visualization
sns.set_style("darkgrid")
# визуализируем
# visualize
sns.lineplot(data=silhouette_df, x="cluster", y="silhouette", marker= "o")
```

```python
# На основе RFM-признаков кластеризуйте клиентов с помощью Agglomerative Clustering
# Cluster customers based on RFM features using Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering

# напишем функцию, чтобы рассчитать коэффициент силуэта
# Let's write a function to calculate the silhouette coefficient
def get_silhouette(cluster_num, X):
    ac = AgglomerativeClustering(n_clusters=cluster_num)
    labels = ac.fit_predict(X)
    # подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
    # calculate the silhouette metric by passing the data and which clusters the objects belong to
    silhouette = silhouette_score(X, labels)
    return silhouette

# создадим пустой словарь, ключами будут коэффициент силуэта и количество кластеров
# create an empty dictionary, the keys will be the silhouette coefficient and the number of clusters
silhouette_res = {"silhouette": [], "cluster": []}

for cluster_num in range(3, 11):
    silhouette_res["silhouette"].append(get_silhouette(cluster_num, X_pca))
    silhouette_res["cluster"].append(cluster_num)
    
# сохраним в датафрейм значение силуэта и количество кластеров
# save the silhouette value and the number of clusters in the dataframe
silhouette_df = pd.DataFrame(silhouette_res)

# установим стиль для визуализиции
# set the style for visualization
sns.set_style("darkgrid")
# визуализируем
# visualize
sns.lineplot(data=silhouette_df, x="cluster", y="silhouette", marker= "o")
```

```python
# На основе RFM-признаков кластеризуйте клиентов с помощью GaussianMixture
# Cluster customers based on RFM features using GaussianMixture
from sklearn.mixture import GaussianMixture

# напишем функцию, чтобы рассчитать коэффициент силуэта
# Let's write a function to calculate the silhouette coefficient
def get_silhouette(cluster_num, X):
    gm = GaussianMixture(n_components=cluster_num, random_state=42)
    labels = gm.fit_predict(X)
    # подсчитаем метрику силуэта, передав данные и то, к каким кластерам относятся объекты
    # calculate the silhouette metric by passing the data and which clusters the objects belong to
    silhouette = silhouette_score(X, labels)
    return silhouette

# создадим пустой словарь, ключами будут коэффициент силуэта и количество кластеров
# create an empty dictionary, the keys will be the silhouette coefficient and the number of clusters
silhouette_res = {"silhouette": [], "cluster": []}

for cluster_num in range(3, 11):
    silhouette_res["silhouette"].append(get_silhouette(cluster_num, X_pca))
    silhouette_res["cluster"].append(cluster_num)
    
# сохраним в датафрейм значение силуэта и количество кластеров
# save the silhouette value and the number of clusters in the dataframe
silhouette_df = pd.DataFrame(silhouette_res)

# установим стиль для визуализиции
# set the style for visualization
sns.set_style("darkgrid")
# визуализируем
# visualize
sns.lineplot(data=silhouette_df, x="cluster", y="silhouette", marker= "o")
```

Next, create a so-called cluster profile. To do this, you need to return from the decomposed data (if you performed dimensionality reduction) to the RFM table (cleaned of outliers).

Group the RFM table by the resulting clusters and calculate the mean for each feature.

To make the results easier to interpret, let's look at one way to visualize a cluster profile: a Radar Chart (polar diagram, or spider web diagram). This is a graphical representation of the values ​​of several equivalent categories in the shape of a web.

A Radar Chart is often used to determine cluster profiles. At the ends of the web, axes are plotted corresponding to the features describing the objects. On each axis, the mean values ​​of the corresponding characteristics are plotted for each cluster. By connecting the points along the axes, we obtain a polygon.

This plot shows a visualization of the features for one of the clusters. It's clear that students in this cluster are more interested in music (Music) and less interested in programming (Programm).

The graph_objects module of the plotly library has a built-in function called Scatterpolar, which allows you to plot a polar plot. Based on this function, we've created the plot_cluster_profile() function, which allows you to visualize the profile of each cluster as a polar plot. It has two parameters: `grouped_data` — the characteristics of objects (clients) grouped into clusters, and `n_clusters` — the number of clusters.

The main requirement for using a polar chart is that all features must be normalized to a single scale, where 1 denotes the maximum and 0 denotes the minimum. We also added a normalization step to the `plot_cluster_profile()` function.

```python
# Функция для визуализации профиля кластеров в виде полярной диаграммы
# Function for visualizing the cluster profile as a polar diagram
def plot_cluster_profile(grouped_data, n_clusters):
    """Функция для визуализации профиля кластеров в виде полярной диаграммы.

    Args:
        grouped_data (DataFrame): таблица, сгруппированная по номерам кластеров с агрегированными характеристиками объектов.
        n_clusters (int): количество кластеров
    """
    # Нормализуем сгруппированные данные, приводя их к масштабу 0-1
    # Normalize the grouped data, bringing it to a scale of 0-1
    scaler = preprocessing.MinMaxScaler()
    grouped_data = pd.DataFrame(scaler.fit_transform(grouped_data), columns=grouped_data.columns)
    # Создаём список признаков
    # Create a list of features
    features = grouped_data.columns
    # Создаём пустую фигуру
    # Create an empty figure
    fig = go.Figure()
    # В цикле визуализируем полярную диаграмму для каждого кластера
    # In a loop, we visualize the polar diagram for each cluster
    for i in range(n_clusters):
        # Создаём полярную диаграмму и добавляем её на общий график
        # Create a polar chart and add it to the overall chart
        fig.add_trace(go.Scatterpolar(
            r=grouped_data.iloc[i].values, # радиусы # radii
            theta=features, # название засечек # serif name
            fill='toself', # заливка многоугольника цветом # fill a polygon with color
            name=f'Cluster {i}', # название — номер кластера # name - cluster number
        ))
    # Обновляем параметры фигуры
    # Update the shape parameters
    fig.update_layout(
        showlegend=True, # отображение легенды # display legend
        autosize=False, # устаналиваем свои размеры графика # set your own chart sizes
        width=800, # ширина (в пикселях) # width (in pixels)
        height=800, # высота (в пикселях) # height (in pixels)
    )
    # Отображаем фигуру
    # Display the figure
    fig.show()
```

Теперь у вас есть удобный инструмент для анализа профиля кластеров. Воспользуйтесь им, чтобы проинтерпретировать результаты, полученные на предыдущем шаге.

They are characterized by high purchase frequency and high total revenue. These customers place orders regularly and generate the most revenue for the company.

**Cluster 1 — inactive or lost customers.**
These customers made purchases a long time ago (high Recency value), have low order frequency, and low total revenue. It's likely that some of them have stopped using the store's services.

**Cluster 2 — active customers with average value.**
These customers made purchases relatively recently and return periodically, but their contribution to revenue is lower than that of customers in the first segment.

These segments can be used to develop marketing strategies: retaining the most valuable customers, encouraging repeat purchases from customers with average activity, and reactivating inactive customers.
